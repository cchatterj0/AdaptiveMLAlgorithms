```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# Data obtained from USP DS Repository https://sites.google.com/view/uspdsrepository
# Load the data in your unique bucket.
# Here AdaptiveMLAlgorithms/Data/gassensor.csv
```

```python
inp_data   = pd.read_csv("../Data/gassensor.csv")
inp_data = inp_data.dropna()
inp_data = inp_data.drop(["Class"], 1)
dataset = (inp_data - inp_data.mean())/inp_data.std()
print(dataset.shape)
print(dataset.head())
```

```python
plt.figure(figsize=(32, 32))
plt.rcParams.update({'font.size': 16})
cols = dataset.columns
for i in range(10): #dataset.shape[1]):
    plt.subplot(7, 2, i+1)
    plt.ylabel(cols[i])
    plt.plot(dataset[cols[i]], lw=2, color='b')
plt.show()
```

------------------------------------------
### Adaptive Linear Eigen Decomposition

```python
# Square root of the correlation matrix
from scipy.linalg import eigh

X = dataset.T
nSamples = dataset.shape[0]

# Compute Correlation matrix and eigen vectors of the generated data
corX = (X @ X.T) / nSamples

# Eigenvalues and eigenvectors of corX
eigvals, eigvecs = eigh(corX)
print("eigvals =\n", np.around(eigvals[::-1], 3))

EstV = np.fliplr(eigvecs)
EstD  = np.diag(eigvals[::-1])
```

```python
# Adaptive algorithm
from numpy import linalg as la

nSamples = dataset.shape[0]
nDim = dataset.shape[1]
nEA = 16

A = np.zeros(shape=(nDim,nDim)) # stores adaptive correlation matrix
N = np.zeros(shape=(1,nEA)) # stores eigen norms
W = 0.1 * np.ones(shape=(nDim,nEA)) # stores adaptive eigenvectors
cos_t1 = np.zeros(shape=(nEA, nSamples)) # cosine error

for iter in range(nSamples):
    cnt = iter + 1
    # Update data correlation matrix A with current sample x
    x = np.array(dataset.iloc[iter])
    x = x.reshape(nDim,1)
    A = A + (1.0/cnt)*((np.dot(x, x.T)) - A)
    etat = 1.0/(500 + cnt)
    # Deflated Gradient Descent
    W = W + etat*(A @ W - 0.5*W @ np.triu(W.T @ A @ W) - 0.5*A @ W @ np.triu(W.T @ W))
    newnorm = la.norm(W, axis=0)
    N = np.vstack([N, newnorm])
    for i in range(nEA):
        u1 = W[:,i]/la.norm(W[:,i])
        u1 = u1.reshape(nDim,1)
        vi = EstV[:,i].reshape(nDim,1)
        cos_t1[i,iter] = abs(u1.T @ vi)
    
print("W=\n", W.shape)
```

```python
import math
from scipy.linalg import eigh
WTW = W.T @ W
WTAW = W.T @ A @ W
print("Diags of WTAW")
print([WTAW[i,i] for i in range(len(WTAW[0]))])
print("Diags of WTW")
print([WTW[i,i] for i in range(len(WTW[0]))])
```

```python
cols = dataset.columns
norms = pd.DataFrame({cols[0]: N[:, 0], cols[1]: N[:, 1], cols[2]: N[:, 2], cols[3]: N[:, 3]})

plt.figure(figsize=(20, 16))
plt.rcParams.update({'font.size': 16})
cols = norms.columns
for i in range(4):
    plt.subplot(4, 1, i+1)
    plt.ylabel('||Phi {}||'.format(i+1))
    plt.plot(norms[cols[i]], lw=2, color='b')
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.rcParams.update({'font.size': 16})

plt.subplot(2, 2, 1)
plt.plot(cos_t1[0,:], lw=2, color='b')
plt.ylabel('Direction Cosine - Component 1')
plt.xlabel('Samples')
plt.subplot(2, 2, 2)
plt.plot(cos_t1[1,:], lw=2, color='b')
plt.ylabel('Direction Cosine - Component 2')
plt.xlabel('Samples')
plt.subplot(2, 2, 3)
plt.plot(cos_t1[2,:], lw=2, color='b')
plt.ylabel('Direction Cosine - Component 3')
plt.xlabel('Samples')
plt.subplot(2, 2, 4)
plt.plot(cos_t1[3,:], lw=2, color='b')
plt.ylabel('Direction Cosine - Component 4')
plt.xlabel('Samples')

plt.show()
```

```python
X = dataset
# Y = (X @ EstV[:,:16] @ EstV[:,:16].T)
Y = (X @ W @ W.T)
Y.columns = dataset.columns
print(Y.shape)
```
