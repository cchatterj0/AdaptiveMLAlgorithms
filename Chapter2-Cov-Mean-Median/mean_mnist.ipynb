```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
```

```python
# Extract the number "8" from keras mnist dataset into array Char8[nDim, nSamples]

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# (unique, counts) = np.unique(y_train, return_counts=True)
# frequencies = np.asarray((unique, counts)).T
# print(frequencies)

num = 8

X1 = np.zeros(shape=(1,784))
for i in range(x_train.shape[0]):
    if (y_train[i] == num):
        X1 = np.vstack([X1, x_train[i].flatten()])
for i in range(x_test.shape[0]):
    if (y_test[i] == num):
        X1 = np.vstack([X1, x_test[i].flatten()])
Char8 = X1.T
Char8 = np.delete(Char8, (0), axis=1)
Char8.shape
```

```python
# Plot original images
# image_array is np.array(num_images, x_size, y_size)

import matplotlib.pyplot as plt

def plot_images(image_array, stt=0, cmap1='gist_gray'):
    fig, axarr = plt.subplots(6, 4, figsize=(8, 12))
    for i in range(0,6):
        for j in range(0,4):
            img = np.reshape(image_array[:,stt+4*i+j],(-1,28))
            axarr[i,j].imshow(img[:,:], cmap=cmap1)
```

```python
#Plot the original character
plot_images(Char8, 0, 'viridis')
```

```python
# Adaptively compute the mean of the character
nSamples = Char8.shape[1]
nDim = Char8.shape[0]
mf1  = np.zeros(shape=(nDim,1)) # stores adaptive mean vector
beta = 1

for iter in range(nSamples):
    x = Char8[:,iter]
    x = x.reshape(nDim,1)

    # Eq.2.4
    mf1 = beta * mf1  + (1.0/(1 + iter)) * (x - beta * mf1)
```

```python
# Use the significant values of the mean
mf2 = np.zeros(shape=(nDim,1))
for i in range(nDim):
    if (mf1[i] > 10):
        mf2[i] = mf1[i]
max(mf2)

# plot the mean
fig, axarr = plt.subplots(1,1, figsize=(8, 12))
img = np.reshape(mf2[:],(-1,28))
axarr.imshow(img[:,:], cmap='viridis')
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
ax = plt.axes(projection='3d')
x = np.linspace(0, 28, 28)
y = np.linspace(0, 28, 28)
X, Y = np.meshgrid(x, y)
ax.contour3D(X, Y, img, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(75, 35)
fig = plt.figure()
ax.plot_surface(X, Y, img, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```

```python
# Plot the character background
W1 = 255-Char8
plot_images(W1,0, 'viridis')
```

```python
# Compute the mean of the character background
nSamples = W1.shape[1]
nDim = W1.shape[0]
mb1  = np.zeros(shape=(nDim,1)) # stores adaptive mean vector
beta = 1

for iter in range(nSamples):
    x = W1[:,iter]
    x = x.reshape(nDim,1)

    # Eq.2.4
    mb1 = beta * mb1  + (1.0/(1 + iter)) * (x - beta * mb1)
```

```python
mb2 = np.zeros(shape=(nDim,1))
for i in range(nDim):
    if (mb1[i] > 10):
        mb2[i] = mb1[i]
max(mb2)

# plot the mean
fig, axarr = plt.subplots(1,1, figsize=(8, 12))
img2 = np.reshape(mb2[:],(-1,28))
axarr.imshow(img2[:,:], cmap='viridis')
```
