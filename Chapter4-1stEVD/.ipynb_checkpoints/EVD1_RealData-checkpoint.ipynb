```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#get data here: https://sites.google.com/view/uspdsrepository
dataset = pd.read_csv("../Data/INSECTS-incremental-abrupt_balanced_norm.csv")
dataset = dataset.dropna()
dataset = dataset.drop('Class', 1) #.iloc[0:10000]
dataset1 = 20*(dataset-dataset.mean())

print(dataset1.shape)
print(dataset1.head())
```

```python
plt.figure(figsize=(32, 48))
plt.rcParams.update({'font.size': 16})
cols = dataset.columns
for i in range(dataset.shape[1]):
    plt.subplot(17, 2, i+1)
    plt.ylabel(cols[i])
    plt.plot(dataset[cols[i]], lw=2, color='b')
plt.show()
```

```python
# Square root of the correlation matrix
from scipy.linalg import eigh

nSamples = dataset1.shape[0]
nDim = dataset1.shape[1]
X = dataset1.T

# Compute Correlation matrix and eigen vectors of the generated data
corX = (X @ X.T) / nSamples

# Eigenvalues and eigenvectors of corX
eigvals, eigvecs = eigh(corX)
print("eigvals =\n", eigvals[::-1])

EstV = np.fliplr(eigvecs)
EstD  = np.diag(eigvals[::-1])
```

```python
# Adaptive algorithm
from numpy import linalg as la

mu = 10
A = np.zeros(shape=(nDim,nDim)) # stores adaptive correlation matrix
w = 0.1 * np.ones(shape=(nDim,11)) # weight vectors of all algorithms
cos_t = np.zeros(shape=(11, nSamples)) # cosine error
Weight = 1

for iter in range(nSamples):

    x = np.array(dataset1.iloc[iter]).reshape(nDim,1)
    A = Weight * A + (1.0/(1 + iter))*((np.dot(x, x.T)) - Weight * A)

    # OJA
    v = w[:,0].reshape(nDim, 1)
    v = v + (1/(100+iter))*(A @ v - v @ (v.T @ A @ v))
    w[:,0] = v.reshape(nDim)

    # OJAN
    v = w[:,1].reshape(nDim, 1)
    v = v + (1/(10+iter))*(A @ v - v @ ((v.T @ A @ v) / (v.T @ v)) )
    w[:,1] = v.reshape(nDim)

    # LUO
    v = w[:,2].reshape(nDim, 1)
    v = v + (1/(100+iter))*(A @ v * (v.T @ v) - v @ (v.T @ A @ v))
    w[:,2] = v.reshape(nDim)

    # RQ
    v = w[:,3].reshape(nDim, 1)
    v = v + (1/(10+iter))*(A @ v - v @ ((v.T @ A @ v) / (v.T @ v)) )
    w[:,3] = v.reshape(nDim)

    # OJA+
    v = w[:,4].reshape(nDim, 1)
    v = v + (1/(10+iter))*(A @ v - v @ (v.T @ A @ v) - v @ (1 - v.T @ v))
    w[:,4] = v.reshape(nDim)

    # IT
    v = w[:,5].reshape(nDim, 1)
    v = v + (4/(1+iter))*((A @ v / (v.T @ A @ v)) - v)
    w[:,5] = v.reshape(nDim)

    # XU
    v = w[:,6].reshape(nDim, 1)
    v = v + (1/(100+iter))*(2 * A @ v - v @ (v.T @ A @ v) - A @ v @ (v.T @ v))
    w[:,6] = v.reshape(nDim)

    # PF
    v = w[:,7].reshape(nDim, 1)
    v = v + (1/(50+iter)) * (A @ v - mu * v @ (v.T @ v - 1))
    w[:,7] = v.reshape(nDim)

    # AL1
    v = w[:,8].reshape(nDim, 1)
    v = v + (1/(100+iter)) * (A @ v - v @ (v.T @ A @ v) - mu * v @ (v.T @ v - 1))
    w[:,8] = v.reshape(nDim)

    # AL2
    v = w[:,9].reshape(nDim, 1)
    v = v + (1/(100+iter))*(2 * A @ v - v @ (v.T @ A @ v) - A @ v @ (v.T @ v) - mu * v @ (v.T @ v - 1))
    w[:,9] = v.reshape(nDim)

    # FENG
    v = w[:,10].reshape(nDim, 1)
    v = v + (1.0/(50+iter))*((v.T @ v) * A @ v - v)
    w[:,10] = v.reshape(nDim)


    v0 = EstV[:,0].reshape(nDim,1)
    for i in range(11):
        u = w[:,i] / la.norm(w[:,i])
        u = u.reshape(nDim,1)
        cos_t[i,iter] = 100*abs(u.T @ v0)
```

```python
plt.figure(figsize=(16, 12))
plt.rcParams.update({'font.size': 16})

plt.subplot(2, 2, 1)
plt.plot(cos_t[0,:], lw=2, color='b', label="OJA")
plt.plot(cos_t[1,:], lw=2, color='r', label="OJAN")
plt.ylabel('Direction Cosines')
plt.xlabel('Samples')
plt.legend(loc="center")

plt.subplot(2, 2, 2)
plt.plot(cos_t[2,:], lw=2, color='b', label="LUO")
plt.plot(cos_t[3,:], lw=2, color='r', label="RQ")
plt.ylabel('Direction Cosines')
plt.xlabel('Samples')
plt.legend(loc="center")

# plt.subplot(3, 2, 3)
# plt.plot(cos_t[4,:], lw=2, color='b', label="OJA+")
# plt.plot(cos_t[5,:], lw=2, color='r', label="IT")
# plt.ylabel('Direction Cosines')
# plt.xlabel('Samples')
# plt.legend(loc="center")

plt.subplot(2, 2, 3)
plt.plot(cos_t[6,:], lw=2, color='b', label="XU")
plt.plot(cos_t[7,:], lw=2, color='r', label="PF")
plt.ylabel('Direction Cosines')
plt.xlabel('Samples')
plt.legend(loc="center")

plt.subplot(2, 2, 4)
plt.plot(cos_t[8,:], lw=2, color='b', label="AL1")
plt.plot(cos_t[9,:], lw=2, color='r', label="AL2")
plt.ylabel('Direction Cosines')
plt.xlabel('Samples')
plt.legend(loc="center")

# plt.subplot(3, 2, 6)
# plt.plot(cos_t[10,:], lw=2, color='b', label="FENG")
# plt.ylabel('Direction Cosines')
# plt.xlabel('Samples')
# plt.legend(loc="center")

plt.show()
```
